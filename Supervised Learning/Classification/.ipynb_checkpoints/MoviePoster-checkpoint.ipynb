# Multi-Label Image Classification Model in Python

In this notebook a machine learning is applied to a multi-label image classification problem. The inputs are Movie Posters and the outputs are the Movie Genres. 

The data set is avaliable for download at the following link: 

- https://drive.google.com/file/d/1iQV5kKF_KGZL9ALx9MMXk_Lg7PklBLCE/view


## <span style="color:navy">Importing libraries</span> 

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.utils import to_categorical
from keras.preprocessing import image
from keras import backend as K
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
%matplotlib inline
```

```python
pip install -U scikit-learn
```

```python
from sklearn.model_selection import train_test_split
```

## <span style="color:navy">Importing the ouputs of the CNN as a Pandas Frame</span> 

```python
train = pd.read_csv('train.csv')    # reading the csv file
train.head()      # printing first five rows of the file
```

```python
train.columns
```

## <span style="color:navy">Importing the Images (Input of CNN)</span> 

```python
pip install pillow
```

```python
train_image = []
for i in tqdm(range(train.shape[0])):
    img = image.load_img('Images/'+train['Id'][i]+'.jpg',target_size=(243,243,3))
    img = image.img_to_array(img)
    img = img/255
    train_image.append(img)
X = np.array(train_image)
```

```python
X.shape
```

```python
plt.imshow(X[2])
```

```python
train['Genre'][2]
```

```python
y = np.array(train.drop(['Id', 'Genre'],axis=1))
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.1)
```

## <span style="color:navy">Building the Convolution Network</span> 

```python
model = Sequential()
model.add(Conv2D(filters=16, kernel_size=(5, 5), activation="relu", input_shape=(243,243,3)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Conv2D(filters=32, kernel_size=(5, 5), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Conv2D(filters=64, kernel_size=(5, 5), activation="relu"))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Conv2D(filters=64, kernel_size=(5, 5), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(25, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

## <span style="color:navy">Training the CNN</span>

```python
model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test), batch_size=64)
```

### <span style="color:navy">Plotting the error distribution of the Test Data Set</span> 

```python
len(X_test)
```

```python
top_cat_predict=np.zeros((726, 3))
```

```python
proba = model.predict(X_test)
```

```python
for d in range (len(X_test)):
    top_cat_predict[d]=(np.argsort(proba[d])[:-4:-1])
```

```python
top_sorted=top_cat_predict
top_sorted.sort()
```

```python
top_cat_actual=100*np.ones((726, 3))
```

```python
for b in range (len(y_test)):
    iterate=y_test[b]
    start=0
    for t in range (len(iterate)):
        if iterate[t]==1:
            top_cat_actual[b][start]=t
            start=start+1
```

```python
top_cat_actual
```

```python
top_sorted
```

```python
errors=np.zeros((25))
```

```python
for outer in range (len(top_sorted)):
    for inner in range (len(top_sorted[0])):
        actual=top_cat_actual[outer][inner]
        predict=top_sorted[outer][inner]
        if actual!=100:
            if actual!=predict:
                errors[int(actual)]=errors[int(actual)]+1              
```

```python
errors
```

```python
plt.figure(figsize=[20,8])
plt.bar(s,errors, color='firebrick')
plt.xticks(rotation=90, fontsize='15', color='gray',fontweight='bold')
plt.yticks(fontsize='15', color='gray',fontweight='bold')
plt.title('Error Distribution of the Test Data Set', fontsize='20', fontweight='bold') 
plt.show()
```

The three Genres with the highest error rate are
- Drama
- Comedy 
- Crime

The following analysis will look at what the drama category of posters are being misclassified as.

### <span style="color:navy">Misclassification distribution for the Drama Errors</span> 

```python
drama_errors=np.zeros((25))
```

```python
for outer in range (len(top_sorted)):
    for inner in range (len(top_sorted[0])):
        actual=int(top_cat_actual[outer][inner])
        predict=top_sorted[outer][inner]
        if actual!=100:
            if actual==7:
                if actual!=predict:
                    drama_errors[int(predict)]=drama_errors[int(predict)]+1 
```

```python
drama_errors
```

```python
plt.figure(figsize=[20,8])
plt.bar(s,drama_errors, color='goldenrod')
plt.xticks(rotation=90, fontsize='15', color='gray',fontweight='bold')
plt.yticks(fontsize='15', color='gray',fontweight='bold')
plt.title('Misclassification Distribution of the Drama category in the Test Data Set', fontsize='20', fontweight='bold') 
plt.show()
```

### Most <span style="color:green">Drama</span> Category Posters are being misclassifed as mainly belonging to <span style="color:red">Comedy, Action or Crime </span> Categories
